Create Mempool file

```python
import sqlite3
import sql_lite

# Connect to the database (replace 'your_database_file.db' with your actual file)
conn = sqlite3.connect('mempool.db')

# Create a cursor object
cursor = conn.cursor()

# Create the 'keys' table if it doesn't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS mempool (
        number INTEGER PRIMARY KEY,
        sender BLOB,
        recipient BLOB,
        value BLOB,
        data BLOB,
        gasLimit BLOB,
        gasPrice BLOB,
        signature BLOB
    )
''')

# Commit the changes to the database
conn.commit()

# Execute the query
cursor.execute('SELECT sender FROM mempool')

# Fetch all the results
results = cursor.fetchall()

# Print or process the results
for row in results:
    print(row)

# Close the connection
conn.close()
```
